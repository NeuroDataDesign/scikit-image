```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.morphology import disk
from skimage.filters.rank import equalize
from skimage.filters.rank import mean_bilateral
from skimage.filters.rank import enhance_contrast
from skimage.filters.rank import otsu
from skimage.filters import sobel
from skimage.filters import roberts
from skimage.filters.rank import noise_filter

from skimage.morphology import closing

from skimage.filters import threshold_local
from skimage.exposure import equalize_hist
```

```python
from generic import equalize_3D
im1 = io.imread("classic.tif")
print(im1.shape)
```

```python
im1 = io.imread("classic.tif")
print(im1.shape)
im1 = np.array(im1[3])


im2 = io.imread("axon in cerebellum.tif")
im2 = np.array(im2[2])

image = np.concatenate([im1, im2])
```

```python
def normalize_pixels(i):
    i = 256*(i/np.amax(i))
    return i.astype(int)
```

```python
print(np.amin(image))

print(image.shape)

bad_mask = np.ones((185, 123), dtype=np.uint8)
print(bad_mask.dtype)

print(bad_mask.shape)
image_equalized = equalize(image, disk(50), mask=bad_mask)

# hist = np.histogram(image, bins=np.arange(10000, 40000))
# plt.figure(2)
# plt.plot(hist[1][:-1], hist[0], lw = 2)

image_equalized = normalize_pixels(image_equalized)
plt.imshow(image_equalized)
```

```python
image_equalized = equalize(image, disk(50), mask=bad_mask)
image_glob_equalized = equalize_hist(image)
bilat_image = mean_bilateral(image, disk(50), s0 = 5, s1 = 5)
enh = enhance_contrast(image, disk(10))
```

```python
## roberts
mage_nf = roberts(image)
image_equalized_nf = roberts(image_equalized)
image_glob_equalized_nf = roberts(image_glob_equalized)
bilat_image_nf = roberts(bilat_image)
enh_nf = roberts(enh)

image_nf = normalize_pixels(image_nf)
image_equalized_nf = normalize_pixels(image_equalized_nf)
image_glob_equalized_nf = normalize_pixels(image_glob_equalized_nf)
bilat_image_nf = normalize_pixels(bilat_image_nf)
enh_nf = normalize_pixels(enh_nf)
```

```python
# Closing

selem = disk(1)
image_cl = closing(image, selem)
image_equalized_cl = closing(image_equalized, selem)
image_glob_equalized_cl = closing(image_glob_equalized, selem)
bilat_image_cl = closing(bilat_image, selem)
enh_cl = closing(enh, selem)

# Closing followed by Otsu

selem = disk(1)
image_cl_ot = otsu(image_cl, selem)
image_equalized_cl_ot = otsu(image_equalized_cl, selem)
image_glob_equalized_cl_ot = otsu(image_glob_equalized_cl, selem)
bilat_image_cl_ot = otsu(bilat_image_cl, selem)
enh_cl_ot = otsu(enh_cl, selem)

# 
image_cl = normalize_pixels(image_cl)
image_equalized_cl = normalize_pixels(image_equalized_cl)
image_glob_equalized_cl = normalize_pixels(image_glob_equalized_cl)
bilat_image_cl = normalize_pixels(bilat_image_cl)
enh_cl = normalize_pixels(enh_cl)

# N cl_ot
image_cl_ot = normalize_pixels(image_cl_ot)
image_equalized_cl_ot = normalize_pixels(image_equalized_cl_ot)
image_glob_equalized_cl_ot = normalize_pixels(image_glob_equalized_cl_ot)
bilat_image_cl_ot = normalize_pixels(bilat_image_cl_ot)
enh_cl_ot = normalize_pixels(enh_cl_ot)
```
